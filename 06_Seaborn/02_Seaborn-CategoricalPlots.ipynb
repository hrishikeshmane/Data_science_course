```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### barplot allows to aggregate the categorical data based on the some function by default it's mean. Visualization of a group by action.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill', data=tips, estimator=np.std)
# standard deviation is a measure that is used to quantify amount of variation or dispersion of a set of data values from the mean.
# A low standard deviation indicates that the data points tend to be close to the mean 
# A high standard deviation indicates that the data points are spread out over a wider range of values.
```

### countplot same as barplot but here estimator is exclusively counting the number of occurences.

```python
sns.countplot(x='sex',data=tips)
```

### boxplot and violinplot used to shown the distribution of the categorical data. boxpot shows the quartile of the datasets and here, dots are outlier. where as violinplot plots the graph for all the data points.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
# A boxplot splits the data set into quartiles. 
# The body of the boxplot consists of a "box", which goes from the first quartile (Q1) to the third quartile (Q3).
# Within the box, a horizontal line is drawn at the Q2, the median of the data set. 
# Two vertical lines, called whiskers, extend from the front and back of the box. 
# The front whisker goes from Q1 to the smallest non-outlier in the data set, 
# and the back whisker goes from Q3 to the largest non-outlier.
# Range - It is the spread of all the data and it is represented on a boxplot by the vertical distance 
# between the smallest value and the largest value, including any outliers.
# Interquartile range (IQR) - The middle half of a data set falls within the interquartile range. 
# In a boxplot, the interquartile range is represented by the width of the box (Q3 minus Q1).
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

### stripplot plots a scatter plot where one variable is categorical

```python
sns.stripplot(x='day',y='total_bill',data=tips) # Result is scatter plot
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True) #Add Random Noise
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

### swarmplot combines scatterplot, striplot with violinplot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

### factorplot is the most general form of all these plots

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind='violin')
```
