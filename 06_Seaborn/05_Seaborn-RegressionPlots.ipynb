### lmplot function allows you to plot linear Models with seaborn.

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips) #linear-fit
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex') # 2 linear fits
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100}) # 's' -> size of the markers
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='day',
           hue='sex', aspect=0.8, size=8)
```
