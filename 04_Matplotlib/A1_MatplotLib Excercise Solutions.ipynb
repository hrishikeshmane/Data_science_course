## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Import matplotlib.pyplot as plt and set %matplotlib inline

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Excercise 1
### Create a figure object called fig using plt.figure()
### use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis as ax
### Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("title")
```

## Excercise 2
### Create a figure object and put two axes on it, ax1 and ax2. 
### Located at [0,0,1,1] and [0.2,0.5,0.2,0.2] respectively.

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

### Now plot (x,y) on both the axes. And call your figure object to show it.

```python
ax1.plot(x,y)
ax2.plot(x,y)
fig
```

## Excercise 3
### Create the plot below by adding the two axes to a figure object at [0,0,1,1] and [0.2,0.5,0.4,0.4]

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

### Now use x,y and z arrays to recreate the plot below. 
### Notice the x limits and y limits on the inserted plot.

```python
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

#Insert plot
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
fig
```

## Excercise 4
### use plt.subplots(nrows=1, ncols=2) to create the plot below.

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

### Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style.

```python
axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3,ls='--')
fig
```

### resize the plot by adding the figsize() argument in plt.subplots()

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3,ls='--')
```
