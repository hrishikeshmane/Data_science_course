# Matplotlib Introduction
* Matplotlib is the most popular plotting library for Python.
* It gives you control over every aspect of a figure or plot.
* It was designed to have a similar feel to MatLab’s graphical plotting.
* For installation using command line or terminal – conda install matplotlib
 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

## Functional Methods

```python
plt.plot(x,y)
```

```python
plt.plot(x,y); # to remove the object address on the terminal
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

## Multi-Plot

```python
plt.subplot(1,2,1) #rows,columns,plot no referring to
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Matplotlib Object Oriented API Methods

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Left, bottom, width and height of the axes
```

```python
axes.plot(x,y)
display(fig)
#fig
```

```python
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
display(fig)
```

## 2 Sets of Figure in one canvas

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title("LARGER PLOT")
axes2.set_title("SMALLER PLOT")
display(fig)
```

## Sub-plot using OO API Methods

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes
```

```python
for current_axes in axes:
    current_axes.plot(x,y)
display(fig)
```

```python
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
display(fig)
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100) #figsize=(width,height), dpi -> dots/pixels per inch
```

```python
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
display(fig)
```

```python
fig,axes = plt.subplots(figsize=(8,2),nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig,axes = plt.subplots(figsize=(8,2),nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

## Saving the figure

```python
fig,axes = plt.subplots(figsize=(8,2),nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
fig.savefig('my_picture.png',dpi=200) # will save in cd, use !pwd to know
```

## Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x ** 2, label="square")
ax.plot(x,x ** 3, label="cube")

ax.legend(loc=0) #Location Code 0 -> Best location
# https://matplotlib.org/api/legend_api.html
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00') #'orange','blue','green','purple' or Hex Code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=5) #linewidth can be used as lw
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=0.5) #alpha parameter controls how transparent line width is
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=3,linestyle='--') #'-.',':','steps' ls is the short form
```

## Markers

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=1,linestyle='--', marker='o',
        markersize=10, markerfacecolor='green',
        markeredgewidth=3, markeredgecolor='black') #marker='+','*','1',
```

## Control over axis appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=3,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```

## Special Plot Types

```python
plt.scatter(x,y) #Scatter Plots
```

```python
from random import sample #Histogram Plot
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)] #Boxplot
plt.boxplot(data, vert=True,patch_artist=True)
```
