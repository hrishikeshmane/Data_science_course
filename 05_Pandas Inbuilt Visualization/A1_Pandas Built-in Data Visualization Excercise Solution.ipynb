```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df3 = pd.read_csv(r"data/df3")
df3.info()
```

```python
df3.head()
```

### Recreate the scatter-plot of b vs a. Note the size and color of the points and also note the figure size.

```python
df3.plot.scatter(x='a',y='b',figsize=(12,3), s=50, c='red')
```

### Create a histogram of the 'a' column

```python
df3['a'].plot.hist(bins=30)
```

### Use Style-sheets to set the style to 'ggplot'.
#### https://matplotlib.org/users/style_sheets.html
#### hint: import matplotlib.pyplot as plt;plt.style.use('ggplot')

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=30,alpha=0.5)
```

### Create a boxplot comparing the a and b columns.

```python
df3[['a','b']].plot.box()
```

### create a kde plot of the 'd' column

```python
df3['d'].plot.kde()
```

### From the Above plot, increase the line-width and make the line-style dashed.

```python
df3['d'].plot.kde(lw=2,ls='--')
```

### Create an area plot of all the columns for just the rows upto 30.(Hint: use .ix)

```python
df3.iloc[0:30].plot.area()
```

### How to display the legend outside of the plot as shown above? Hint: https://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas

```python
f = plt.figure()
df3.iloc[0:30].plot.area()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
